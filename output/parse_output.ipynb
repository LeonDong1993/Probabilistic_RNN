```python
# parse training time
import numpy as np
from collections import *
def parse_minute(x):
    x = list(map(int,x))
    return (x[0]*60 + x[1] + x[2]/60)
content = open('train_time.txt').readlines()
res = defaultdict(list)
for item in content:
    if 'train time for 50 trial' in item:
        line = item.strip().split(':')
        name = line[1].split(' ')[0]
        res[name].append(parse_minute(line[-3:]))
        
tmp = res
res = []
order = ['NN-MG','RNN-STD','RNN-IndMGx3', 'RNN-MG']
for i in range(11):
    for x in order:
        res.append(tmp[x][i])
        
        

res = np.array(res).reshape(-1,4).T
for row in res:
    for item in row:
        print(item, end= ' ')
    print('')

```
